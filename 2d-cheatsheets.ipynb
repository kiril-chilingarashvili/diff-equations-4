## L7.2. Block diagrams.

### Engineers build complex systems by combining simpler ones. This process is recorded by the symbolism of ***block diagrams***.

### A system acts on the input signal to produce the system response. To be precise, one should specify initial conditions as well, but
### 1. These are often implied by an assumption – of steady state or of zero initial conditions; and
### 2. They do not matter much anyway if the system is stable.

### So they are ignored in the block diagram notation.
### In the time domain we can represent the input signal $y$, system $S$, and the system response $x$, by the following diagram.
![img](img/diag1.png)

### We can convert this to the frequency domain by applying Laplace transform to the variables $x$ and $y$. What happens to the system? It gets represented by its transfer function:
![img](img/diag2.png)

### In the $s$-domain,
## $$ {\color{orange}{X(s)}} = H(s) {\color{blue}{Y(s)}} $$
### In the frequency domain, the effect of the system is simply to multiply by the transfer function.

### This makes it easy to explain what happens if you “cascade" two system: feed a second system (with transfer function $G(s)$, say), as input signal, the output signal of a first system (with transfer function $H(s)$):
![img](img/diag3.png)

### The cascade is another LTI system, with its own transfer function, and we have just seen that the transfer function of this new system is just the product $G(s)H(s)$.

### This observation has the following amazing consequence: The order in which you cascade is immaterial! Feeding the output from system 1 into system 2 has the same effect as feeding the output of system 2 into system 1. And both of these are equivalent to the the following block diagram with a single block whose entry is the product of the two cascaded blocks.
![img](img/diag4.png)

### That is
## $$ X(s) = G(s) (H(s)Y(s)) \quad \text{and} \quad X(s) = H(s) (G(s)Y(s)) $$
### or equivalently,
## $$ X(s) = (H(s) G(s))Y(s) $$

## L7.4. Block diagrams are not circuit diagrams.

### ***Warning***: 
### Circuit diagrams and block diagrams have a superficial resemblance. But they are different! The state of a circuit diagram (current through various components and voltage drop across them) varies with time. The wires carry electrons. In contrast, the lines in a block diagram are labeled with entire signals, known for all time. We need the full signal, because the system components represented by the blocks operate on these signals (differentiating them, for example).

### In a block diagram the input signal generally comes in from the left, and the system response exits from the right. In contrast a circuit diagram requires a loop, a ***circuit***. The input signal and system response need to be identified separately.

### A “series" relationship in a block diagram is a cascade; ***the transfer functions multiply***.

## L7.5. Impedance of resistors.

### We can use the language of transfer functions and complex gain to describe and understand impedance in electric circuits.

### ***Ohm's law and linear resistance***.
### A “linear" resistor produces a particularly simple relationship between the current flowing through it and the voltage drop across it:
## $$ V_R = R I_R $$
![img](img/c10.png)
### If we embed this electrical component in a circuit driven by a voltage $V$ we have a simple system. We'll take the input to be $V$ and the output to be $I$, the current across $V$.
![img](img/c11.png)

### For such a simple system we have $V=V_R$ and $I=I_R$. So $V=R I$ becomes $\displaystyle I = \frac{1}{R}$. That is, the transfer function is the constant $\displaystyle \frac{1}{R}$. the reciprocal of the resistance.

### The effects of the other electronic components we consider – the capacitor and the inductance coil – are more complicated, but when we consider them in the frequency domain they are similar enough to treat in a way exactly parallel with the story of resistance, which we recall below. The generalization of resistance is called impedance. Any configuration of resistors, inductors, and capacitors placed in a circuit with a (possibly time varying) driving voltage has an impedance. The impedance is not necessarily a constant. Rather it is a function in the frequency domain, which we will see is the reciprocal of the transfer function.

### ***Impedance of resistors***.
### Impedance is useful in analyzing circuits with components arranged in series. We use the reciprocal of impedance to analyze components in parallel. You can see this already in the case of resistors.
### Putting two resistors (with resistances $R_1$ and $R_2$, say) in series results in a system equivalent to a single resistor with constant $R$.
![img](img/c12.png) 

### To see what $R$ is, just observe that the current through everything is the same, $I$ say; so the voltage drop across the first is $R_1 I$, and across the second is $R_2 I$: for a total of $(R_1+R_2)I$. So
## $$ R = R_1+R_2 $$
### This signals a general fact: Impedances of components connected in series add.

### What if we put them in parallel? Again we get a system that is equivalent to a single resistor.
![img](img/c13.png) 

### The voltage drop across each of them is the same, while the current is split between the two branches:
## $$ I_R = I_1+I_2,\quad V_R=R_1 I=R_2 I $$
### This turns out to be equivalent to a single resistor. To see what the total resistance is, put the two equations together:
## $$ I_R = \frac{V_R}{R_1} + \frac{V_R}{R_2} = V_R(\frac{1}{R_1} + \frac{1}{R_2}) $$
### or
## $$ V_R = R I_r $$
### with
## $$ \frac{1}{R} = \frac{1}{R_1} + \frac{1}{R_2} $$
### This signals a general fact: The reciprocals of impedances of components in parallel add, and is equal to the reciprocal of the total impedance.

## L7.6. Impedance of capacitors and inductors.

### The rules relating the voltage drop to the current are:
### - Capacitor: $C\dot{V}_C = I_C$
### - Inductor: $V_L = L\dot{I}_L$
### Here, $V_L, I_L, V_C, I_C$ are analogous to $V_R$ and $I_R$ across a resistor.
![img](img/c20.png)

### To convert these differential relations into ratios, we will now move into the frequency domain by applying the Laplace transform. So
## $$ \mathcal{I}(s) = \mathcal{L} (I(t); s), \quad \mathcal{V}_R (s) = \mathcal{L} (V_R(t);s),\quad\mathcal{V}_C(s)=\mathcal{L} (V_C(t);s),\quad\mathcal{V}_L(s)\mathcal{L} (V_L(t);s) $$
### Our convention of capitalization to indicate formation of the Laplace transform isn't available, since our variables are already capitalized, so we will indicate it by use of script letters. (This problem is the reason engineers like to use $j$ for $\sqrt{-1}$: then $i$ can be current, and $I$ its Laplace transform. Now, why “$i$" for “current" is a different question ….)

### Then (assuming rest initial conditions), the Laplace transform of the different voltage rules are:
## $$ \mathcal{V}_R=R \mathcal{I}_R, \quad \mathcal{V}_C = \frac{1}{sC}\mathcal{I}_C,\quad \mathcal{V}_L = sL \mathcal{I}_L $$
### As you can see, the role of “resistance" $R$ is played by $\displaystyle \frac{1}{Cs}$ in the case of a capacitor and by $Ls$ in the case of an inductor. The Laplace transform of voltage drop is a multiple of the transformed current, but the multiplier is no longer constant but rather depends upon $s$. This multiplier is the reciprocal of the system function, as we will see in the example below. It is called the ***impedance*** or ***complex impedance*** of the component.

### ***Comment***: 
### As usual, $s=i\omega$ is an important physical case and you will sometimes see the impedance defined in terms of $i\omega$ instead of $s$.

### ***Transfer functions of single component systems***
### We saw above that a circuit with a single resistor has transfer function $\displaystyle \frac{1}{R}$. Similarly, if we have a circuit with, say, a single inductor
![img](img/c21.png)

### For such a simple circuit, it's easy to see that $V=V_L$ and $I=I_L$. Thus, the voltage law for inductors says that
## $$ L\dot{I} = V $$
 
### If we call $V$ the input and $I$ the output of this system then clearly the system function is $\displaystyle \frac{1}{sL}$, which is the reciprocal of the impedance of the inductor.

### Likewise, the system function for a circuit with a single capacitor is $Cs$ which is the reciprocal of the impedance of the capacitor.

## L7.7. Impedance of components in series.

### ***Example 7.1***
### In a ***series*** RLC circuit the current is the same everywhere, $I=I_R=I_C=I_L$, and the voltage boost by the power source equals the sum of the voltage drops across the components. In the frequency domain,
## $$ \mathcal{V}=\mathcal{V_L}+\mathcal{V_R}+\mathcal{V_C}=Ls\mathcal{I}+R\mathcal{I}+\frac{1}{sC}\mathcal{I} = (Ls+R+\frac{1}{sC})\mathcal{I} $$
![img](img/c30.png)

### If we declare the input signal to be the voltage increase across the power source and the system response to be the current through the circuit, the transfer function is
## $$ \frac{\mathcal{I}}{\mathcal{V}} = \frac{1}{Ls+R+\frac{1}{sC}} = \frac{s}{Ls^2+Rs+\frac{1}{C}} $$
### a result we have seen before. But now the ***reciprocal of the transfer function***, the impedance,
## $$ Ls+R+\frac{1}{sC} $$
### is revealed to be an analogue of the resistance.

### The impedance of this series is the sum of the impedances of the three constituent components, just as for resistances. The corresponding impedances are
### - Inductor: $Ls$
### - Resistor: $R$
### - Capacitor: $\displaystyle \frac{1}{C}$
